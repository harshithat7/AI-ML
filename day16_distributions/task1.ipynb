```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
heights = np.random.normal(170, 10, 1000)
incomes = np.random.exponential(50000, 1000)
scores = np.random.normal(80, 10, 1000) 
df = pd.DataFrame({
    'Heights': heights,
    'Incomes': incomes,
    'Scores': scores
})
plt.figure(figsize=(12,4))
for i, col in enumerate(df.columns):
    plt.subplot(1, 3, i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
for col in df.columns:
    mean = df[col].mean()
    median = df[col].median()
    print(f"{col}: Mean={mean: .2f}, Median={median: .2f}")
    if mean > median:
        print("Right-Skewed")
    elif mean < median:
        print("Left-Skewed")
    else:
        print("Symmetric")
```
