```python
import pandas as pd
import numpy as np 
data = {'Score': [50, 60, 70, 80, 90, 100, 150, 200]}
df = pd.DataFrame(data) 
mu = df['Score'].mean()
sigma = df['Score'].std() 
df['z_score'] = (df['Score'] - mu) / sigma 
outliers = df[np.abs(df['z_score']) > 3]
print(df)
print("\nOutliers:")
print(outliers)
```
