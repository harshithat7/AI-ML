```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
data = {'Salary': [50000, 60000, 40000, 70000, 55000],
        'Age': [25, 30, 22, 35, 28]}
df = pd.DataFrame(data)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.hist(df['Salary'], label='Original Salary')
plt.title('Original')
scaler = StandardScaler()
df_std = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
plt.subplot(1, 3, 2)
plt.hist(df_std['Salary'], label='Standardized Salary', color='g')
plt.title('Standardized')
scaler = MinMaxScaler()
df_norm = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
plt.subplot(1, 3, 3)
plt.hist(df_norm['Salary'], label='Normalized Salary', color='r')
plt.title('Normalized')
plt.tight_layout()
plt.show()
```
