```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25, 36, 49, 64])
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
print("Original R^2:", r2_score(y, y_pred))
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model.fit(X_poly, y)
y_pred_poly = model.predict(X_poly)
print("Polynomial R^2:", r2_score(y, y_pred_poly))
plt.scatter(X, y, label='Data')
plt.plot(X, y_pred, label='Linear', color='g')
plt.plot(X, y_pred_poly, label='Polynomial', color='r')
plt.legend()
plt.show()
```
