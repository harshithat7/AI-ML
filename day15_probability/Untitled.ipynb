```python
import pandas as pd
df = pd.read_csv(r"C:\Users\user\Downloads\car_price_prediction.csv")
print(df.head())
print(df.info())
print(df.isnull().sum())
df["Levy"]=pd.to_numeric(df["Levy"],errors="coerce")
df["Levy"].fillna(df["Levy"].median() ,inplace = True)
df.dropna(subset=['Price', 'Prod. year'], inplace=True)
df['Manufacturer'].fillna('Unknown', inplace=True)
df['Price'] = df['Price'].astype(int)
df['Prod. year'] = df['Prod. year'].astype(int)
df['Mileage'] = df['Mileage'].astype(str)
df['Cylinders'] = df['Cylinders'].astype(float)
df['Manufacturer'] = df['Manufacturer'].str.strip().str.upper()
df['Model'] = df['Model'].str.strip().str.title()
df.drop_duplicates(subset=['ID'], inplace=True)
df = df[df['Prod. year'] <= 2024]
df = pd.get_dummies(df, columns=['Category', 'Fuel type'], drop_first=True)
df.to_csv(r"C:\Users\user\Downloads\car_price_prediction.csv", index=False)
```
