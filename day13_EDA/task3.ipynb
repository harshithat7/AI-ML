```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {'Price': [200,300,400,500,600,700,800,1000,1200],
        'SquareFootage': [1000,1500,1200,2000,1800,2500,2200,3000,3500],
        'Bedrooms': [2,3,2,4,3,4,3,5,5]}
df = pd.DataFrame(data)
plt.figure(figsize=(8,6))
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
print("High correlation (>0.8):")
print(corr[(corr > 0.8) & (corr != 1)])
plt.figure()
sns.boxplot(x=df['Price'])
plt.title('Price Outliners')
plt.show()
```
