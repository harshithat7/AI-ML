```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {'Price': [200,300,400,500,600,700,800,1000,1200],
        'SquareFootage': [1000,1500,1200,2000,1800,2500,2200,3000,3500],
        'City': ['NYC', 'LA', 'NYC', 'Chicago', 'LA', 'NYC', 'Chicago', 'LA', 'NYC']}
df = pd.DataFrame(data)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.scatterplot(x='SquareFootage', y='Price', data=df)
plt.title('SquareFootage vs Price')
plt.subplot(1,2,2)
sns.boxplot(x='City', y='Price', data=df)
plt.title('Price by City')
plt.tight_layout()
plt.show()
```
