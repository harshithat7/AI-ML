```python
import pandas as pd
data = {'Price' : ['$10.5', '$20', '$15.75'],
        'Date' : ['2026-01-01', '2026-01-02', '2026-01-03']}
df = pd.DataFrame(data)
print("Initial dtypes:")
print(df.dtypes)
df['Price'] = df['Price'].str.replace('$', '').astype(float)
df['Date'] = pd.to_datetime(df['Date'])
print("\nUpdated dtypes:")
print(df.dtypes)
print("\nDataFrame:")
print(df)
```
