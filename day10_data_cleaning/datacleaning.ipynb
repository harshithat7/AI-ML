```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

```python
data = pd.read_csv(r"C:\Users\user\Downloads\sample2.csv")
data.head()
```

```python
data.drop('Unnamed: 0', axis=1, inplace = True)
```

```python
data.describe()
```

```python
fig, axs = plt.subplots(1,4, figsize = (16,6))
axs[0].boxplot(data.Age)
axs[1].boxplot(data.Salary)
axs[2].boxplot(data.Experience)
axs[3].boxplot(data.Monthly_Sales)
plt.show()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
```

```python
IQR
```

```python
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR
```

```python
upper_limit, lower_limit
```

```python
outliners = ((data < lower_limit) | (data > upper_limit)).any(axis=1)
outliner_data = data[outliners]
cleaned_data = data[~outliners]
```

```python
outliner_data
```

```python
cleaned_data
```

```python
cleaned_data.head()
```

```python
fig, axs = plt.subplots(1,4, figsize = (16,6))
axs[0].boxplot(cleaned_data.Age)
axs[1].boxplot(cleaned_data.Salary)
axs[2].boxplot(cleaned_data.Experience)
axs[3].boxplot(cleaned_data.Monthly_Sales)
plt.show()
```

```python
z_score = np.absolute(zscore(data))
z_score_mask = (z_score > 3).any(axis=1)
```
